# Intro

Some plotting examples specifically for working in ipython notebooks.

```python
#fancy inline plots, this will load numpy and matplotlib functions and 
# methods into the local name space
%pylab nbagg
```

```python
# create interval of equally spaced numbers and print it:
x = linspace(1,10,10) 
x
```

```python
# open figure and make a line plot:
ifig=1; close(ifig); figure(ifig)
plot(x,sin(x))
```

```python
ifig=2; close(ifig); figure(ifig)
bar(x-0.5,sin(x),width=1.0)
```

```python
f =lambda x: x**2
ifig=3; close(ifig); figure(ifig)
plot(x,f(x),'g-o',label='$f(x)$')
legend(loc=0)
xlabel('x')
ylabel('$ f(x) = x^2 $')
```
